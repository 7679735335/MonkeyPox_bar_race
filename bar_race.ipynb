```python
import bar_chart_race as bcr
import pandas as pd
```

```python
df=pd.read_csv("/content/Monkeypox_confirmed_cases_Dataset.csv")
df
```

```python
df.dtypes
```

```python
df[["Date"]] = df[["Date"]].apply(pd.to_datetime)
```

```python
df.set_index('Date', inplace=True)
```

```python
for name in df.columns:
  #print(name)
  df[name]=df[name].cumsum()
```

```python
df.tail()
```

```python
bcr.bar_chart_race(df,period_length=1000,title='MonkeyPox Confirmed Cases by Country',n_bars=8,period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total cases: {v.nlargest(6).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'})
```
